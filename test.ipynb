```python
COLLECTION_NAME = "ncert-vectors"
from sentence_transformers import SentenceTransformer
from qdrant_client import QdrantClient, models
# Ensure the embedding model is the same one used for uploading the vectors
EMBEDDING_MODEL_NAME = 'all-MiniLM-L6-v2'
VECTOR_SIZE = 384 # Should match your uploaded vectors

# --- 1. Initialize Qdrant Client and Embedding Model ---
print("Initializing Qdrant client for Qdrant Cloud...")
try:
    client = QdrantClient(
        url="https://1bcdfef2-22c6-44f9-a2fa-adc4a4ca5e4e.europe-west3-0.gcp.cloud.qdrant.io:6333",
        api_key="eyJhbGciOiJIUzI1NiIsInR5cCI6IkpXVCJ9.eyJhY2Nlc3MiOiJtIn0.erRscERUav-m7n0_T5FbZkvCFakbTMlcuM5Rqu-Q038",
    )
    # Test connection
    cluster_info = client.cluster_info()
    print(f"Connected to Qdrant Cloud. Cluster status: {cluster_info.status}")
    collection_info = client.get_collection(collection_name=COLLECTION_NAME)
    print(f"Collection '{COLLECTION_NAME}' exists. Points count: {collection_info.points_count}")

except Exception as e:
    print(f"Error connecting to Qdrant Cloud or accessing collection: {e}")
    print("Please ensure your Qdrant Cloud URL/API Key are correct and the collection exists.")
    exit()

print(f"Loading embedding model: {EMBEDDING_MODEL_NAME}...")
try:
    embedding_model = SentenceTransformer(EMBEDDING_MODEL_NAME)
    print("Embedding model loaded.")
except Exception as e:
    print(f"Error loading embedding model: {e}")
    print("Please ensure 'sentence-transformers' is installed and the model name is correct.")
    exit()

# --- 2. Query Function ---

def run_query(query_text, top_k=5, filter_criteria=None):
    """
    Performs a semantic search query against the Qdrant Cloud collection.

    Args:
        query_text (str): The natural language query string.
        top_k (int): The number of top relevant results to return.
        filter_criteria (dict, optional): A dictionary of metadata filters.
                                           e.g., {"subject": "Physics", "class": "Class 11"}
    """
    print(f"\n--- Querying for: '{query_text}' ---")

    # Generate the vector embedding for the query text
    query_vector = embedding_model.encode(query_text).tolist()

    search_params = {
        "collection_name": COLLECTION_NAME,
        "query_vector": query_vector,
        "limit": top_k, # How many results to return
        "with_payload": True, # Return the payload (metadata + content)
        "with_vectors": False # No need to return the vectors themselves for typical search display
    }

    if filter_criteria:
        # Construct Qdrant filter from the dictionary
        qdrant_filter = models.Filter(
            must=[
                models.FieldCondition(
                    key=key,
                    match=models.MatchValue(value=value)
                ) for key, value in filter_criteria.items()
            ]
        )
        search_params["query_filter"] = qdrant_filter

    try:
        hits = client.search(**search_params)

        if not hits:
            print("No results found for this query or filter combination.")
            return

        print(f"Found {len(hits)} results:")
        for i, hit in enumerate(hits):
            print(f"\nResult {i+1} (Score: {hit.score:.4f}):")

            # Safely get payload values with default 'N/A'
            # Assuming your payload has 'content' and then 'class', 'subject', 'chapter', 'page'
            # based on your SQLite upload script.
            class_val = hit.payload.get('class', 'N/A')
            subject_val = hit.payload.get('subject', 'N/A')
            chapter_val = hit.payload.get('chapter', 'N/A')
            page_val = hit.payload.get('page', 'N/A')
            full_content_val = hit.payload.get('content', 'N/A') # Your payload stores it as 'content'

            print(f"  Class: {class_val}")
            print(f"  Subject: {subject_val}")
            print(f"  Chapter: {chapter_val}")
            print(f"  Page: {page_val}")
            print(f"  Content Snippet: {full_content_val[:300]}...") # Show first 300 chars

    except Exception as e:
        print(f"An error occurred during search: {e}")
        print("Please check your Qdrant Cloud connection, collection name, and ensure payload keys match.")

# --- Example Queries ---

# Basic semantic search (general query across all subjects)
run_query("third law of motion. To every action, there is always an equal andopposite reaction.")
```
